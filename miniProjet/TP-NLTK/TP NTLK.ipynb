```python
# (a)

def freqByLetters(iterable):
    textbyletters = []
    letters = set()
    freqLetters = dict()

    for w in iterable:
        for l in w:
            textbyletters.append(l)
    letters = set(textbyletters)

    freqDist = FreqDist(textbyletters)

    for l in letters:
        freqLetters[l]=freqDist.freq(l)

    return freqLetters

print(freqByLetters(text5))
print(freqByLetters(text6))
```
